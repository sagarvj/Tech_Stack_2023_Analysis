```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('survey_results_public.csv')
```

```python
#LanguagesHaveWorkedWith
languages = df['LanguageHaveWorkedWith'].str.split(';', expand= True)
languages.stack().value_counts()
```

```python
language_counts = languages.stack().value_counts().head(10).sort_values(ascending= True)

language_counts.plot(kind = 'barh', color = sns.color_palette("Blues",len(language_counts)),figsize = (32,16))

plt.ylabel('Programming Languages', fontsize=30, fontweight = 'bold')
plt.xlabel('Number of Users',fontsize=30, fontweight = 'bold')
plt.title('Top 10 Programming Languages',fontsize=40, fontweight = 'bold')

#label values for each bar
for index, value in enumerate(language_counts):
    plt.text(value * 1.003 , index, f'{value}', ha ='left', fontsize=23, fontweight = 'bold', va='center')

# x&y labels with fontsize 
plt.xticks(fontsize=25)
plt.yticks(fontsize=25,fontweight = 'bold')

#labelling reference text (bottom right corner)
max_x=language_counts.max()/2
offset = 0.45 * language_counts.max()
plt.text(max_x + offset, 0.5, 'Total Responses: 87140',fontsize=20,fontweight= 'bold',bbox=dict(facecolor='lightblue',alpha=0.2),ha='center')

#display barchart
plt.tight_layout()
plt.savefig('top_languages.png')

plt.show()
```
